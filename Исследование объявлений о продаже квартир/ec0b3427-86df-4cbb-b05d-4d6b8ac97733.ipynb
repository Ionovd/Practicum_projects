```python
def full_column_info(df):
    values = pd.DataFrame()
    
    for i in df.columns:
        values[i] = [df[i].unique(), df[i].dtypes, df[i].nunique()]
    return values.transpose().sort_values(by=1)

full_column_info(df)  
```

```python
def mis_info(df):
    count_mis = df.isna().sum()
    percent_mis = round(df.isna().sum() * 100 / len(df), 2)
    
    mis_info_df = pd.DataFrame({'name': df.columns,
                                'count_mis': count_mis,
                                '%': percent_mis})
    
    return mis_info_df.sort_values(by= '%', ascending=False)
mis_info(df)
```

Высота потолка в 100 метров и 1 метр вызывают сомнения, проверим столбец полностью и заменим пропуски медианными значениями.

```python
# Заменим все значения NaN на медианные по кол-ву этажей.
replace_m = df.groupby('floors_total')['ceiling_height'].median()
for i in df['floors_total'].unique():
    df.loc[(df['floors_total'] == i) & (df['ceiling_height'].isna()), 'ceiling_height'] = replace_m[i]
```

```python
# Исключим потолки выше 10м и ниже 2м, сочтем это как ошибку, на общую выборку влияния не будет.
df.loc[df['ceiling_height'] < 2,'ceiling_height'] = None
df.loc[df['ceiling_height'] > 10,'ceiling_height'] = None
df = df.dropna(subset=['ceiling_height'])
```

```python
# Создадим функцию для дальнейшей экономии времени, группировка и замена медианными значениями.
def median_fill_gr (nan_column, gr_column):
    for value in df[gr_column].unique():
        median_gr_column = df.loc[df[gr_column]==value, nan_column].median()
        df.loc[(df[nan_column].isna()) & (df[gr_column]==value), nan_column] = median_gr_column
# Создадим функцию для медианных значений без группировки.
def median_fill (nan_column):
    df.loc[df[nan_column].isna() ,nan_column] = df[nan_column].median()
```
